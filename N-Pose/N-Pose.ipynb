```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from transforms3d.euler import quat2euler

# Load the data
data = pd.read_csv('L Forearm - 10_D422CD006143_20240528_160631.csv', skiprows=6)

# Assuming SampleTimeFine is in milliseconds, convert it to seconds from the start of the recording
data['TimeInSeconds'] = (data['SampleTimeFine'] - data['SampleTimeFine'].min()) / 1000000

# Convert quaternion to Euler angles (radians and then to degrees)
# Note: quat2euler returns angles in the order of (roll, pitch, yaw)
data['Euler_X'], data['Euler_Y'], data['Euler_Z'] = zip(*data.apply(
    lambda row: quat2euler([row['Quat_W'], row['Quat_X'], row['Quat_Y'], row['Quat_Z']]), axis=1))
data['Euler_X'] = np.degrees(data['Euler_X'])
data['Euler_Y'] = np.degrees(data['Euler_Y'])
data['Euler_Z'] = np.degrees(data['Euler_Z'])

# Set up the figure and subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(14, 10))
fig.suptitle('Left Arm N-Pose')

# Plot Euler angles
axes[0].plot(data['TimeInSeconds'], data['Euler_X'], label='Euler X (Roll)')
axes[0].plot(data['TimeInSeconds'], data['Euler_Y'], label='Euler Y (Pitch)')
axes[0].plot(data['TimeInSeconds'], data['Euler_Z'], label='Euler Z (Yaw)')
axes[0].set_title('Euler Angles Over Time')
axes[0].set_xlabel('Time (seconds)')
axes[0].set_ylabel('Angle (degrees)')
axes[0].legend()

# Plot acceleration
axes[1].plot(data['TimeInSeconds'], data['Acc_X'], label='Acc X')
axes[1].plot(data['TimeInSeconds'], data['Acc_Y'], label='Acc Y')
axes[1].plot(data['TimeInSeconds'], data['Acc_Z'], label='Acc Z')
axes[1].set_title('Acceleration Over Time')
axes[1].set_xlabel('Time (seconds)')
axes[1].set_ylabel('Acceleration (m/s)')
axes[1].legend()

# Display the plots
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
